```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
FILE = "data.csv"
```

```python
def fun(a: np.float64, b: np.float64, c: np.float64) -> np.float64:
    return a**2 + 2*a*b + b**2 + c
```

```python
def vis(proj_idx: int) -> None:
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')

    x = np.linspace(0, 10, 100)
    y = np.linspace(0, 10, 100)
    X, Y = np.meshgrid(x, y)
    Z = np.zeros(shape=X.shape)
    for i in range(X.shape[0]):
        for j in range(Y.shape[1]):
            values = [0.0] * 3
            values[(proj_idx + 1) % 3] = X[0][i]
            values[(proj_idx + 2) % 3] = Y[j][0]
            Z[i][j] = fun(*values)
    ax.plot_surface(X, Y, Z)
    ax.set_title("Projection {proj}".format(proj=proj_idx))
    plt.show()
```

```python
vis(proj_idx=0)
```

```python
vis(proj_idx=1)
```

```python
vis(proj_idx=2)
```
